```python
import tf2onnx
import onnxruntime as rt
from tensorflow import keras
from core import *
import numpy as np
from sklearn.metrics import mean_absolute_error
import tensorflow as tf
```

```python
#Load Neural Network model and data
model = load_model()
df = load_data()
scaler = load_scaler()
```

```python
sample_data = df.sample(frac=0.05)
X = sample_data.drop(columns=["trip_duration_seconds"]).to_numpy()
X = scaler.transform(X)
Y = sample_data["trip_duration_seconds"].to_numpy()
pred = np.expm1(model.predict(X))
MAE = mean_absolute_error(Y, pred)
print(MAE)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

with open("taxi_model.tflite", "wb") as f:
    f.write(tflite_model)
```
