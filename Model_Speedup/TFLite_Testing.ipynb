```python
import numpy as np
import tensorflow as tf
from core import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# load the tflite model
model = load_TF_model()
scaler = load_scaler()
data = load_data()
```

```python
#Sample some data

# X = data.drop(columns=["trip_duration_seconds"]).to_numpy()
# X = scaler.transform(X)
# Y = data["trip_duration_seconds"].to_numpy()
# X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
# X_scaled = scaler.transform(X_test).astype('float32')

sample_data = data.sample(frac=0.05)
X = sample_data.drop(columns=["trip_duration_seconds"]).to_numpy()
X = scaler.transform(X)
Y = sample_data["trip_duration_seconds"].to_numpy()

X_scaled = X.astype('float32')
```

```python
#Get input/output details
input_details = model.get_input_details()
output_details = model.get_output_details()

model.resize_tensor_input(input_details[0]['index'], [len(X_scaled), 7])
model.allocate_tensors()

model.set_tensor(input_details[0]['index'], X_scaled)
model.invoke()
predictions = np.expm1(model.get_tensor(output_details[0]['index']))
```

```python
MAE = mean_absolute_error(Y, predictions)
print(MAE)
```
