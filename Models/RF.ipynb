```python
from core import *
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.utils import resample
from sklearn.metrics import mean_absolute_error
```

```python
df=load_data()
```

```python
#Split data into testing and training
X = df[['VendorID', 'passenger_count', 'trip_distance', 'payment_type', 
        'flag_no_meter', 'flag_overcharge', 'pickup_hour', 'pickup_dayofweek', 
        'pickup_month', 'is_weekend', 'is_rush_hour', 'store_fwd_flag_int', 
        'ratecode_clean']]
y = df['trip_duration_seconds']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_sample, y_sample = resample(X_train, y_train, n_samples=100000, random_state=42)
```

```python
#Train the random forest
rf = RandomForestRegressor()
rf.fit(X_sample,y_sample)

#Show which features are important for decision making
importances = pd.Series(rf.feature_importances_, index=range(X_train.shape[1]))
print(importances.sort_values(ascending=False))
```

```python
rf.score(X_test, y_test)
```

```python
y_pred = rf.predict(X_test)
MSE = mean_absolute_error(y_test, y_pred)
print(f"MSE: {MSE}")
```
